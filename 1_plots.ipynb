# - Histograms, boxplots, correlation

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

<!-- Could add: Bar chart, grouped bar chart, Line plot, Pie chart, violin plots -->

```python
data = np.random.randn(1000)

plt.boxplot(data)

# plt.xlabel('Values')
# plt.ylabel('Frequency')
plt.title('Boxplot using Matplotlib')

plt.show()
```

```python
sns.histplot(data, bins=20, color='red')

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram using Seaborn')

plt.show()
```

```python
data_1 = np.array([5, 6, 7, 8, 9])
data_2 = np.array([1, 2, 3, 4, 5])

plt.scatter(data_1, data_2)

plt.xlabel('data_1')
plt.ylabel('data_2')
plt.title('Scatter plot using Matplotlib')

plt.show()
```

```python

features = {
    "feature_1": [1, 2, 3, 4, 5, 6, 7],
    "feature_2": [2, 3, 5, 7, 13, 15, 17],
    "feature_3": [5, 6, 7, 11, 15, 17, 19],
    "feature_4": [3, 4, 11, 13, 17, 21, 23]
}
df = pd.DataFrame(features)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True)
```
